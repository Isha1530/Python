```python
# Tuple Operations
# 1. Create tuples with different data types
integer_tuple = (1, 2, 3, 4, 5)
float_tuple = (1.1, 2.2, 3.3, 4.4)
string_tuple = ("apple", "banana", "cherry")
mixed_tuple = (1, "apple", 3.14, True)

# 2. Access tuple elements using positive and negative indices
print(integer_tuple[0])  # Positive index
print(integer_tuple[-1])  # Negative index

# 3. Perform tuple slicing
tuple_slice = integer_tuple[1:4]
print(tuple_slice)

# 4. Count occurrences and find the index of an element
print(mixed_tuple.count("apple"))
print(mixed_tuple.index(3.14))

# 5. Use built-in functions with tuples
print(len(integer_tuple))
print(max(float_tuple))
print(min(float_tuple))
print(sum(integer_tuple))

# 6. Count and print distinct elements from a tuple
def count_distinct_elements(t):
    return len(set(t))

print(count_distinct_elements(mixed_tuple))

# 7. Convert a list to a tuple and vice versa
list_to_tuple = tuple([1, 2, 3])
tuple_to_list = list(("a", "b", "c"))
print(list_to_tuple)
print(tuple_to_list)

# 8. Demonstrate unpacking of tuples
a, b, c, d = mixed_tuple
print(a, b, c, d)

# List Operations
# 1. Create a list of integers, strings, and mixed data types
integer_list = [1, 2, 3, 4, 5]
string_list = ["apple", "banana", "cherry"]
mixed_list = [1, "apple", 3.14, True]

# 2. Access elements, perform slicing, and update list elements
print(integer_list[0])
print(string_list[-1])
list_slice = integer_list[1:3]
print(list_slice)
integer_list[0] = 10
print(integer_list)

# 3. Add and remove elements
integer_list.append(6)
integer_list.insert(1, 15)
integer_list.remove(3)
integer_list.pop()
print(integer_list)

# 4. Concatenate and repeat lists
concatenated_list = integer_list + string_list
repeated_list = string_list * 2
print(concatenated_list)
print(repeated_list)

# 5. Create a list of squares of the first 10 natural numbers
squares = [x**2 for x in range(1, 11)]
print(squares)

# 6. Filter even numbers using list comprehension
evens = [x for x in range(1, 11) if x % 2 == 0]
print(evens)

# 7. Sorting, reversing, and copying lists
sorted_list = sorted(integer_list)
reversed_list = list(reversed(integer_list))
list_copy = integer_list[:]
print(sorted_list)
print(reversed_list)
print(list_copy)

# 8. Remove duplicates from a list
def remove_duplicates(lst):
    return list(set(lst))

print(remove_duplicates([1, 2, 2, 3, 4, 4, 5]))

# Dictionary Operations
# 1. Create a dictionary
my_dict = {"name": "Alice", "age": 25, "city": "New York"}

# 2. Access, update, and delete elements
print(my_dict["name"])
my_dict["age"] = 26
print(my_dict)
del my_dict["city"]
print(my_dict)

# 3. Use dictionary methods
print(my_dict.keys())
print(my_dict.values())
print(my_dict.items())

# 4. Add and remove key-value pairs
my_dict["country"] = "USA"
del my_dict["age"]
print(my_dict)

# 5. Create a nested dictionary
student_details = {
    "student1": {"name": "John", "age": 20, "marks": 85},
    "student2": {"name": "Emma", "age": 22, "marks": 90}
}

# 6. Access and update nested dictionary elements
print(student_details["student1"]["name"])
student_details["student1"]["marks"] = 88
print(student_details)

# 7. Merge two dictionaries
dict1 = {"a": 1, "b": 2}
dict2 = {"c": 3, "d": 4}
dict1.update(dict2)
print(dict1)

# 8. Sort a dictionary based on values
sorted_dict = dict(sorted(dict1.items(), key=lambda item: item[1]))
print(sorted_dict)

# Set Operations
# 1. Create a set
my_set = {1, 2, 3, 4, 5}

# 2. Add and remove elements
my_set.add(6)
my_set.remove(3)
print(my_set)

# 3. Combine elements from two sets
set1 = {1, 2, 3}
set2 = {3, 4, 5}
union_set = set1 | set2
print(union_set)

# 4. Find common elements
intersection = set1 & set2
print(intersection)

# 5. Find elements in one set but not another
difference = set1 - set2
print(difference)

# 6. Elements in either set but not in their intersection
symmetric_difference = set1 ^ set2
print(symmetric_difference)

# 7. Check subset and superset
print(set1.issubset(union_set))
print(union_set.issuperset(set1))

# 8. Remove all elements
my_set.clear()
print(my_set)
```
