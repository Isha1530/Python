```python
# heatmap
```

```python
f_df = sns.load_dataset('flights')
f_df.head()
```

```python
# Pivot the DataFrame
f_df = f_df.pivot(index="month", columns="year", values="passengers")

# Display the top 5 rows
f_df.head()

```

```python
plt.figure(figsize=(12,6))
ax =sns.heatmap(f_df)
ax
```

```python
plt.figure(figsize=(12,6))
ax= sns.heatmap(f_df,annot=True,fmt='d')
```

```python
plt.figure(figsize=(12,6))
ax= sns.heatmap(f_df,annot=True,fmt='d',linewidths=5,linecolor='b')
```

```python
plt.figure(figsize=(12,6))
ax= sns.heatmap(f_df,annot=True,fmt='d',linewidths=5,linecolor='k',cmap = 'icefire')
```

```python
grplt.figure(figsize=(12,6))
sns.heatmap(f_df,cbar = False)
```

```python
# Set up the grid keyword arguments correctly
grid_kws = {'height_ratios': (0.4, 0.05), 'hspace': 0.4}

# Create subplots with gridspec_kw
fig, (ax, cbar_ax) = plt.subplots(nrows=2, ncols=1, gridspec_kw=grid_kws, figsize=(8, 6))

# Create heatmap
sns.heatmap(f_df, cbar_kws={'orientation': 'horizontal'}, ax=ax, cbar_ax=cbar_ax)

# Show plot
plt.show()
```

```python
# Load Titanic dataset
t_df = sns.load_dataset('titanic')

# Compute correlation matrix (numeric columns only)
corr_matrix = t_df.corr(numeric_only=True)

# Plot heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

# Show the plot
plt.show()
```
