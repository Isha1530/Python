```python
sujects = ['maths','physics','chemistry','computer','english']
marks1 = [10,20,30,40,50]
marks2 = [5,15,20,20,10]
```

```python
plt.figure(figsize=(8,8))
plt.bar(sujects,marks1,color='red')
plt.bar(sujects,marks2,color='blue')
plt.xlabel('subjects')
plt.ylabel('marks')
plt.show()
```

```python
suject_len = np.arange(len(sujects))
width = 0.4
```

```python
plt.figure(figsize=(8,8))

plt.bar(suject_len ,marks1,color='red',width=width)
plt.bar(suject_len +width ,marks2,color='blue',width=width)
plt.xlabel('subjects')
plt.ylabel('marks')
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.bar(suject_len ,marks1,color='red',width=0.4)
plt.bar(suject_len +width ,marks2,color='blue',width=0.4,alpha=0.7)
plt.xlabel('subjects')
plt.ylabel('marks')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example data (Replace with actual data)
mark1 = [45, 50, 60, 70, 85, 90, 95]
mark2 = [40, 55, 65, 75, 80, 88, 92]

# Define bins
bins = np.arange(0, 105, 5)  # Ensure the range covers all marks

# Create histogram
plt.figure(figsize=(8, 8))
plt.hist([mark1, mark2], bins=bins, label=['Mark1', 'Mark2'], alpha=0.7, edgecolor='black')

# Fix axis labels
plt.xticks(np.arange(0, 105, 5))
plt.xlabel('Marks')  # Corrected typo
plt.ylabel('Frequency')
plt.title('Histogram of Marks Distribution')
plt.legend()  # Add legend for clarity

# Show plot
plt.show()
```

```python
plt.figure(figsize=(8,8))

plt.subplot(2,2,1)
plt.scatter(std_hours,marks)

plt.subplot(2,2,2)
plt.plot(std_hours,marks,'r--')

plt.subplot(2,2,3)
plt.hist(marks,label=std_hours)

plt.subplot(2,2,4)
plt.plot(std_hours,marks,'r--')
plt.plot(std_hours,marks,'bo')

plt.savefig(r'C:\Users\HP\Downloads\subplot.png', facecolor='k')
```
